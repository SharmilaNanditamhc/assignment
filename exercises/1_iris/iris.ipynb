# Iris Dataset Analysis

## Prepare the dataset

Use `numpy` or to load `data/iris.data`. Remember to look at the file contents to understand the data format.

```python
import numpy as np  
import matplotlib 
data_set = np.genfromtxt("iris.data", delimiter=",", usecols=(0, 1, 2, 3))
  

```

Create a map that returns the attribute name given the attribute index.

- 0 -> sepal length
- 1 -> sepal width
- 2 -> petal length
- 3 -> petal width

```python
attribute_map = {
    0: "sepal length",
    1: "sepal width",
    2: "petal length",
    3: "petal width"
}
```

## Data analysis

Write a function that prints the mean and another that prints the standard deviation of each attribute.

```python
def print_means(data):
    means = np.mean(data, axis=0)
    for i, val in enumerate(means):
        print(f"Mean of {attribute_map[i]}: {val:.2f}")

def print_stds(data):
    stds = np.std(data, axis=0)
    for i, val in enumerate(stds):
        print(f"Standard Deviation of {attribute_map[i]}: {val:.2f}")
```

Use the functions you defined above to print the mean of the attributes for each class.

```python
class_labels = np.genfromtxt("iris.data", delimiter=",", usecols=4, dtype=str)
np.unique(class_labels)
```

Now, let's plot this data. Use `matplotlib` to create scatter plots for petal and sepal. Use different colors for each class.

```python
unique_classes = np.unique(class_labels)

for c in unique_classes:
    class_data = data_set[class_labels == c]
    print(f"\nClass: {c}")
    print_means(class_data)
    print_stds(class_data)
```

```python
import matplotlib.pyplot as plt

# Define colors for classes
colors = {
    "Iris-setosa": "red",
    "Iris-versicolor": "green",
    "Iris-virginica": "blue"
}
```

```python
for c in unique_classes:
    class_data = data_set[class_labels == c]
    plt.scatter(class_data[:, 2], class_data[:, 3], label=c, color=colors[c])

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Width")
plt.legend()
plt.show()
```

```python
for c in unique_classes:
    class_data = data_set[class_labels == c]
    plt.scatter(class_data[:, 0], class_data[:, 1], label=c, color=colors[c])

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Width")
plt.legend()
plt.show()
```
